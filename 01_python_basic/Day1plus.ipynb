## 6-3. 정규표현식

- 정규표현식(Regular Expression)은 문자열을 처리할 때 특정 패턴을 정의하는데 사용되는 문자열
- 문자열 패턴을 사용하여, 문자열 검색, 대체, 추출 작업 등을 수행
- 정규표현식은 특정 규칙을 가지고있어, 다양한 문자와 연산자를 조합하여 복잡한 문자 패턴을 표현
- 크롤링 또는 자연어처리에서 데이터의 특정 패턴을 찾거나 추출할 때 사용

**메타 문자 (Meta Characters)**
- 문자가 가진 원래의 의미가 아닌 특별한 용도로 사용되는 문자
- 정규표현식에서는 대괄호 [] 안에 포함된 문자들 중 하나와 매치를 뜻함

```python
# . ^ $ +
import re
```

```python
# 대괄호 []: 대괄호 내 포함된 문자들 중 하나와 매칭되는지 확인
pattern = r'[abc]'
# Row string -> r 문자열 내에 메타 문자가 포함되어 있다
```

```python
text1 = 'apple is 사과'
re.findall(pattern, text1)
```

```python
pattern = r'[a-z]' # [ ] 내 두 문자 사이에 - 기호를 사용하면, 두 문자 사이의 범위를 지정
re.findall(pattern, text1)
# findall(): 문자열 전체에서 패턴과 일치하는 모든 부분을 찾아 리스트로 변환
```

```python
pattern = r'[가-힣]' # 한글 유니코드의 자음 + 모음 조합 범위
text2 = "so you don't cry for me 세월 지나도 난 변하지 않아 ~"
re.findall(pattern, text2)
```

```python
pattern = r'[가-힣]+' # 공백을 기준으로 문자의 패턴을 확인
re.findall(pattern, text2)
```

```python
# ^ : 해당 문자열을 제외한 나머지 패턴을 확인
pattern = r'[^가-힣]+' 
re.findall(pattern, text2)
```

```python
pattern = r'[a-zA-Z]+' # 모든 알파벳 표현 단어를 추출
re.findall(pattern,text2)
```

```python
pattern = r'\d+' # \d: 모든 숫자/ 
text3 = "1초라도 안보이면, 2이렇게 초조한데, 3삼초는 어떻게 기다려 이아아 2020"
re.findall(pattern,text3)
```

```python
# 소괄호는 특정 패턴 앞이나 뒤의 값을 확인
pattern = r'(\d+)\원'
text4= '2000원짜리 아이스크림 4개 주세요. 500원 잔돈입니다.'
re.findall(pattern,text4)
```

```python
# re.sub() : 패턴과 일치하는 모든 부분을 다른 문자로 대치
text5 = "그댄 나의 챔피언 너와 나의 챔피언 우리 함께 외치면 승리하리라 한국"
pattern = r"챔피언"
replace_pattern = '챔피온~'

re.sub(pattern, replace_pattern, text5)
```

```python
# re.split(): 특정 패턴을 기준으로 문자를 분할
pattern = r'\W+'
re.split(pattern, text5)
```

```python
# 데이터 추출 (Filter)/ loc[(조건식)] : 특정 조건(True)에 해당하는 데이터를 추출
# 층별화: 범주형 데이터 항목의 특성에 따라 다른 데이터의 통계적 특성이 달라질 수 있다.

# 월렌탈비용이 10만원 이상인 고객들 
cond1 = (df1['Amount_Month'] >= 100000) # 월렌탈비용이 10만원 이상인 조건
df1.loc[cond1] # 해당 조건에 대한 데이터를 추출
```

```python
# 월렌탈비용이 10만원 이상이면서 
# 연체건수가 5만원 이하인 고객명단을 추출
cond1 = (df1['Amount_Month']>=100000)
cond2 = (df1['Overdue_count'] <=5)
df1.loc[cond1&cond2] # 두 조건을 모두 만족하는 경우 (And &)
```

```python
df1.loc[cond1|cond2] # 두 조건 중 하나의 조건만 만족해도 데이터를 불러옴 (OR |)
```

```python
# 판매유형(Sales_Type)이 '렌탈'에 해당하는 데이터만 추출
cond1 = (df1['Sales_Type']=='렌탈')
df1.loc[cond1]
```

```python
# 유입경로(channel)이 '영업방판' '홈쇼핑/방송' 대형마트A
# 여러 항목에 대해 데이터 추출
cond1 = (df1['Channel'].isin(['영업방판','홈쇼핑/방송','대형마트A']))
df1.loc[cond1]
```

```python
# 결제은행이 현대카드에 해당하는 계약 명단 중
# 월렌탈비용이 1500000원 이상만 상위 50명의 계약명단을 추출하여
#result3.xlsx 에 저장
cond1 = (df1['Bank']=='현대카드')
cond2 =  (df1['Amount_Month']>=150000)
df1.loc[cond1&cond2].head(50).to_excel('result3.xlsx')
```
