# 4. 함수

## 4-1. 함수의 개념

- 함수: 특정 기능을 수행하는 코드의 집합

```python
# 함수의 생성
def func1(x):
    return x + 200
```

```python
# 입력 매개변수 없이 함수를 구성
def func5():
    return 2000
```

```python
# 입력값이 몇 개가 될지 모르는 함수를 구성해야 할 때
def func7(*x): # * : 여러개의 자료를 한 번에 **Packing**하여 받아들임
    print(type(x))
    return sum(x)
```

```python
# 특정 입력값의 초기값 세팅이 가능
def func8(x1, x2, x3=30):
    return x1 + x2 + x3
```

```python
# 덧셈을 수행하지만, 필요할 땐 뺄셈을 수행해주는 함수를 구성
def func9(x1, x2, x3=False):
    if x3 == True:
        return x1 - x2
    else:
        return x1 + x2
```

- **재귀함수** (recursion): 함수 내에서 자기 함수 스스로가 참조되어 작용하는 함수
- 기저 조건 (Base Case): 재귀함수 내에서 자기 자신이 호출되지 않도록 하는 조건
- 기저 조건이 따로 없다면 무한재귀로 함수가 지속적으로 수행

```python
def func_recurs1(x):
    # 기저 조건 (x값이 반복되며 더해지다가, 10이 되는 순간 재귀가 정지
    if x == 10:
        print('End')
    else: 
        print(x)
        func_recurs1(x+1)
```

```python
# 람다 함수: 함수를 한 줄로 간결하게 만들 때 사용
# 함수의 결과를 변수로 선언하여 활용

def func20(a,b):
    return a+b
```

## 4-4. 클래스 Class
- 특정 함수를 반복하여 사용할 때, 함수의 결과값을 계속 유지한 채로 계산할 수 있음
- 함수를 여러개의 변수로 선언하여 손쉽게 사용 -> 클래스
- 프로그래밍(개발) 에서는 클래스가 많이 사용됨
- 라이브러리 개념을 이해하기 위해 설명

```python
def func30(x):
    global result 
    #  glabal 특정 변수가 함수내에서만 선언되지 않고, 전체 코드에서 사용가능하게 함
    result += x
    return result
```

```python
class add_class():
    # 초기화 함수를 설정
    # 특정 값의 합을 계산하는 변수를 0으로 초기화
    # 앞서 함수에서 사용한 global 역할 수행
    def __init__ (self):
        self.result = 0
        # self: 클래스를 변수로 선언하여 사용할 때,
        # 변수명에 자동으로 클래스 내의 함수를 전달하기 위함
        
    def func40(self, x):
        self.result += x
        return self.result
```
