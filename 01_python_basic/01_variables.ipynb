# 2. 파이썬 기초 문법 

## 2-1. 변수

- 변수: 데이터를 들고다니는 공간
- 변수를 선언한다: 특정 값이나 결과를 사용자가 지정한 이름으로 사용한다 (오른쪽의 값을 등호기호 = 를 기준으로 왼쪽에 사용자가 지정한 이름으로 선언)

## 2-2. 데이터 타입

- 파이썬에서는 데이터 타입: 숫자형 / 문자형 / 논리형 (참/ 거짓)
- 숫자형: 정수형(int)/ 실수형(float)
- 문자형: 문자열(str)/ 범주형(object)
- 논리형: 참/ 거짓(True/ False// bool) > 첫글자를 대문자로 사용
- 파이썬은 대/소문자를 명확히 구분
- type(): 특정 변수안에 있는 데이터 타입을 확인
- 주석 (Markdown): 파이썬 코드에 영향을 주지않는 설명을 입력하는 공간

## 2-4. 파이썬 자료 구조

- 하나의 변수에 여러개의 데이터를 들고 다니고 싶을 때
- 데이터를 들고 다닐 수 있는 상자의 종류

      - List (리스트)
      - Tuple (튜플)
      - Set (셋, 집합)
      - Dictionary (딕셔너리, 사전)

```python
# 리스트 [ ] 대괄호를 사용하여 데이터를 적재
# 데이터상 순서가 존재 (입력 순서를 기억)/ 중복값 허용
# 새로운 데이터를 추가, 삭제, 변경
data1 = [20,30,40]
```

```python
# 기존 리스트에 들어가있는 해당 값을 찾아 삭제
data1.remove(30)

# 특정 위치에 특정 값을 새로 추가
data1.insert(0,300)

# Indexing 색인: 특정 위치에 존재하는 데이터를 찾는 것/ 대괄호 활용
# 맨 끝에서 위치를 계산해 데이터를 찾을 때
# Indexing에서 음수 기호는 맨 뒤에서부터 순서를 부여
data1[-1]

# 특정 범위의 데이터를 모두 확인하고자 할 때
data1[0:1]
```

```python
# 튜플 (Tuple) 소괄호를 이용해 선언
# 데이터의 순서가 존재/ 중복값 허용
# 새로운 데이터를 추가, 삭제, 변경하는 것이 불가능
data2 = (90, 100, 20, 100, 15)
```

```python
# 셋 (set/ 집합) 중괄호를 이용해 선언
# 데이터 상 순서가 존재하지 않음/ 중복 허용 X
# 데이터를 추가, 삭제, 변경 가능
# Indexing 불가
# 집합 연산 가능
# & : 교집합 연산
# | : 합집합 연산
# - : 차집합 연산
# ^ : 대상 차집합
```

## 2-5. 문자열 처리

- 문자 데이터에 대한 다양한 처리 기법
- 정형 데이터뿐만 아니라, 자연어 처리/ 크롤링 등 다양하게 사용

```python
# 따옴표 활용
# len (): 특정 자료의 데이터 개수를 확인
# \n : 줄바뀜 기호
# '''    ''' : 줄바꿈도 허용
# 문자열에서 print 함수를 사용할 땐, 줄바꿈/ tab/ 들여쓰기 등이 포함됨
```

```python
# 포멧팅
# input (): 외부에서 사용자가 직접 값을 입력하는 함수
mbti = input('MBTI를 입력하시오')

f'안녕하세요 제 MBTI는 {mbti}입니다.'
```
