## 6-3. 정규표현식

- 정규표현식(Regular Expression)은 문자열을 처리할 때 특정 패턴을 정의하는데 사용되는 문자열
- 문자열 패턴을 사용하여, 문자열 검색, 대체, 추출 작업 등을 수행
- 정규표현식은 특정 규칙을 가지고있어, 다양한 문자와 연산자를 조합하여 복잡한 문자 패턴을 표현
- 크롤링 또는 자연어처리에서 데이터의 특정 패턴을 찾거나 추출할 때 사용

**메타 문자 (Meta Characters)**
- 문자가 가진 원래의 의미가 아닌 특별한 용도로 사용되는 문자
- 정규표현식에서는 대괄호 [] 안에 포함된 문자들 중 하나와 매치를 뜻함

```python
# . ^ $ +
import re

# 대괄호 []: 대괄호 내 포함된 문자들 중 하나와 매칭되는지 확인
pattern = r'[abc]'
# Row string -> r 문자열 내에 메타 문자가 포함되어 있다
```

```python
text1 = 'apple is 사과'
re.findall(pattern, text1)

pattern = r'[a-z]' # [ ] 내 두 문자 사이에 - 기호를 사용하면, 두 문자 사이의 범위를 지정
re.findall(pattern, text1)
# findall(): 문자열 전체에서 패턴과 일치하는 모든 부분을 찾아 리스트로 변환

pattern = r'[가-힣]' # 한글 유니코드의 자음 + 모음 조합 범위
text2 = "so you don't cry for me 세월 지나도 난 변하지 않아 ~"
re.findall(pattern, text2)
```

```python
# 공백을 기준으로 문자의 패턴을 확인
pattern = r'[가-힣]+'

# ^ : 해당 문자열을 제외한 나머지 패턴을 확인
pattern = r'[^가-힣]+' 

# 모든 알파벳 표현 단어를 추출
pattern = r'[a-zA-Z]+' 

# \d: 모든 숫자/ 
pattern = r'\d+' 

# 소괄호는 특정 패턴 앞이나 뒤의 값을 확인
pattern = r'(\d+)\원'
re.findall(pattern,   )
```

```python
# re.sub() : 패턴과 일치하는 모든 부분을 다른 문자로 대치
text5 = "그댄 나의 챔피언 너와 나의 챔피언 우리 함께 외치면 승리하리라 한국"
pattern = r"챔피언"
replace_pattern = '챔피온~'

re.sub(pattern, replace_pattern, text5)
```

```python
# re.split(): 특정 패턴을 기준으로 문자를 분할
pattern = r'\W+'
re.split(pattern, text5)
```
