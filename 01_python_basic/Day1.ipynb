## 3-1. 조건문 (if)

```python
data1 = 10

# 특정 비교구문이나 조건을 걸어, 해당 조건이 True 값이 나오는 경우, 아래의 종속문장을 수행
if data1 >= 30 : 
    # 들여쓰기: tab 키나 space bar 키를 이용하여 공백을 주어 문장의 종속관계를 표현
    print('해당 데이터는 30보다 크다')
else: # 위의 조건문이 맞지 않는 경우 (False) 아래의 종속문장이 수행
    print('해당 데이터는 30보다 작다')
```

```python
# 여러 조건을 걸어 해당 조건에 맞는 코드를 수행하고자 할 때
data2 = 75

# 여러 조건이 걸렸을 때, 순차적으로 맨 위 구문부터 확인하면서 수행
if data2 >= 90:
    print('A')

#앞서 if에 있는 비교구문이 True가 나오지 않았을 때, elif의 비교구문을 확인
elif data2 >= 80:
    print('B') 
elif data2 >= 70:
    print('C')
else :
    print('D')
```

## 3-2. 반복문 (While/ For)

```python
# while

data7 = 200
while data7 >= 150: # while문 옆 비교구문이 True 경우, 아래의 종속문장을 반복
    print('Go')
    break # 앞서 반복을 수행하다 break 구문을 만나게 되면 반복이 중단
```

```python
data8 = 0

while data8 < 10:
    print(data8)
    #data8 = data8 + 1 # 기존의 변수에 특정 값을 더해 Update
    data8 +=1 # 같은 의미임
```

```python
# 조건 초기화
data9 = 0
while data9 < 5:
    data9 += 1
    if data9 ==3:
        print('A')
        continue 
        # while 문 뒤에 존재하는 모든 종속문장을 무시하고, 비교구문으로 돌아가 반복을 재수행
    print(data9)
```

```python
# For 반복문: 특정 범위나 횟수만큼 종속문장을 반복
# range(n,m): 특정 범위의 함수를 출력하는 구문 (n번째부터, m번 이전까지 정수를 출력)
for i in range(1,10): # range에서 생성된 정수가 하나씩 1 매개변수에 선언이 되어 반복
    # i 매개변수: 특정 구문 내에서 사용하는 변수
    print(i)
```

```python
#리스트 내 데이터 갯수만큼 for문을 반복
#리스트 내 값을 순서대로 하나씩 매개변수에 선언하여 반복
for i in [100,200,300,400]:
    print(i)
```

## 3-3. 예외 처리 구문 (Try/ Except/ Finally)

- 특정 구문을 수행하다가 구조적인 오류들이 발생
- 오류를 방지하기 위해, 사용하는 예외 처리 구문
- 프로그램 형태로 파이썬 코드를 수행시킬 때 많이 사용
- 데이터 분석에서는 크롤링 또는 업무자동화(RPA)같은 자동화된 반복 수행 작업에 필수로 사용

```python
#수행하고자 하는 구문
try : 
    data14 = []
    for i in range(0,10):
        data14.append(i)
    data14.remove(20)
# try 구문이 정상적으로 실행되다가, 오류가 발생할 경우 아래 종속문장을 수행
except Exception as e: 
    # Exception: 예외처리 구문이 작동되다가, 해당 오류 메시지를 출력
    #as e: 해당 구문(Exception)을 축약하여 지정한 매개변수(e) 사용
    print("e")
# 앞서 모든 구문이 끝나고 최종적으로 작동하는 구문
finally:
    print(data14)
```

- **재귀함수** (recursion): 함수 내에서 자기 함수 스스로가 참조되어 작용하는 함수
- 기저 조건 (Base Case): 재귀함수 내에서 자기 자신이 호출되지 않도록 하는 조건
- 기저 조건이 따로 없다면 무한재귀로 함수가 지속적으로 수행

```python
def func_recurs1(x):
    # 기저 조건 (x값이 반복되며 더해지다가, 10이 되는 순간 재귀가 정지
    if x == 10:
        print('End')
    else: 
        print(x)
        func_recurs1(x+1)
```

```python
func_recurs1(1)
```

- 함수를 여러개의 변수로 선언하여 손쉽게 사용 -> 클래스
- 프로그래밍(개발) 에서는 클래스가 많이 사용됨
- 라이브러리 개념을 이해하기 위해 설명

```python
def func30(x):
    global result 
    #  glabal 특정 변수가 함수내에서만 선언되지 않고, 전체 코드에서 사용가능하게 함
    result += x
    return result
```

```python
class add_class():
    # 초기화 함수를 설정
    # 특정 값의 합을 계산하는 변수를 0으로 초기화
    # 앞서 함수에서 사용한 global 역할 수행
    def __init__ (self):
        self.result = 0
        # self: 클래스를 변수로 선언하여 사용할 때,
        # 변수명에 자동으로 클래스 내의 함수를 전달하기 위함
        
    def func40(self, x):
        self.result += x
        return self.result

```

```python
funcA = add_class()
funcB = add_class()
```

```python
funcA.func40(100)
funcA.func40(200)
funcA.func40(500)
```

```python
funcB.func40(0)
```

```python
funcB.func40(200)
```

```python
funcB.func40(50)
```
