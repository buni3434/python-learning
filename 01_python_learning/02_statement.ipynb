# 3. 파이썬 구문 (Statement)

- 특정 코드를 반복하거나, 특정 조건에 대해 원하는 코드가 실행될 수 있도록 구성
- 조건문: 특정 조건에 대해 코드가 실행될 수 있도록 구성
- 반복문: 특정 범위나 횟수에 의해 코드가 반복될 수 있도록 구성
- 예외구문: 코드를 실행함에 있어, 예외상황에 대한 동작을 따로 수행하도록 구성

## 3-1. 조건문 (if)

```python
# 특정 비교구문이나 조건을 걸어, 해당 조건이 True 값이 나오는 경우, 
# 아래의 종속문장을 수행

data1 = 10

# data1이 30보다 크면 print ('해당 데이터는 30보다 크다') 수행
if data1 >= 30 : 
    print('해당 데이터는 30보다 크다')
else: # 위의 조건문이 맞지 않는 경우 (False) 아래의 종속문장이 수행
    print('해당 데이터는 30보다 작다')
```

```python
# 여로 조건을 걸어 해당 조건에 맞는 코드를 수행하고자 할 때,
# 순차적으로 맨 위 구문부터 확인하면서 수행
# 앞서 if에 있는 비교구문이 True가 나오지 않았을 때, elif의 비교구문을 확인
# 문자열도 가능

data2 = 75
if data2 > 90:
    print('A')
elif data2 > 80:
    print('B')
else:
    print('C')
```

```python
# 리스트 내 특정 데이터가 있는지 판별
# in 구문을 기준으로 옆에 있는 값이 위에 있는 자료구조에 포함되어있는지 판별
data3 = [20,30,40,50,10]
if 100 in data3:
    print('A')
else:
    print('B')
```

```python
# 리스트 내 특정 값이 없는지 확인
if 100 not in data3:
    print('A')
else:
    print('B')
```

```python
# 조건문 표현식: 조건문을 더 짧고 간결하게 표현
data5 = 200
if data5 >= 50:
    print('up')
else:
    print('down')

'up' if data5 >= 50 else 'down'
```

## 3-2. 반복문 (while/ For)

```python
# while
# while 문 옆 비교구문이 True인 경우 아래의 종속문장을 반복
# 앞서 반복을 수행하다 break 구문을 만나게 되면 반복이 중단

data6 = 200
while data6 >= 200:
    print('GO')
    break
```

```python
data7 = 0
while data7 < 10:
    print(data7)
    data7 +=1 # 기존의 변수에 특정 값을 더해 update
```

```python
# 조건 초기화
data8 = 0
while data8 < 5:
    data8 += 1
    if data8 == 3:
        print('A')
        continue
        # while 문 뒤에 존재하는 모든 종속문장을 무시하고, 비교구문으로 돌아가 반복을 재수행
    print(data8)
```

```python
# For 반복문: 특정 범위나 횟수만큼 종속문장을 반복
# range(n,m): 특정 범위의 함수를 출력하는 구문 (n번째부터, m번 이전까지 정수를 출력)
for i in range(1,10): # range에서 생성된 정수가 하나씩 1 매개변수에 선언이 되어 반복
    # i 매개변수: 특정 구문 내에서 사용하는 변수
    print(i)
```

```python
#리스트 내 데이터 갯수만큼 for문을 반복
#리스트 내 값을 순서대로 하나씩 매개변수에 선언하여 반복
for i in [100,200,300,400]:
    print(i)
```

```python
# Index: 데이터의 순서 (Indexing 특정 순서의 데이터를 확인한다)
# 특정 리스트의 데이터를 Index(순서)와 함께 출력
data10 = ['홍길동','성춘향','이몽룡','허준','이순신']
for i in range(len(data10)):
    print(i, data10[i])
# Series 순열: Index(순서) - value(데이터 값)으로 구성된 데이터 구조
```

```python
# 리스트 내 앞/뒤 순서의 자료에 대한 연산을 동시에 수행
data11 = [200,300,400,500,500]

for i in range(len(data11)-1): # 총 4회 반복 수행
    print(data11[i] + data11[i+1])
```

```python
# List Comprehension 리스트 컴프리헨션
# For문을 이용해서 축약한 뒤 특정 리스트를 구성, 해당 리스트를 축약해서 사용
data13 = []
for i in range(1,11):
    data13.append(i)
data13
```

```python
[x for x in range(1,11)]
```

## 3-3. 예외 처리 구문 (Try/ Except/ Finally)

- 특정 구문을 수행하다가 구조적인 오류들이 발생
- 오류를 방지하기 위해, 사용하는 예외 처리 구문
- 프로그램 형태로 파이썬 코드를 수행시킬 때 많이 사용
- 데이터 분석에서는 크롤링 또는 업무자동화(RPA)같은 자동화된 반복 수행 작업에 필수로 사용

```python
#수행하고자 하는 구문
try : 
    data14 = []
    for i in range(0,10):
        data14.append(i)
    data14.remove(20)
# try 구문이 정상적으로 실행되다가, 오류가 발생할 경우 아래 종속문장을 수행
except Exception as e: 
    # Exception: 예외처리 구문이 작동되다가, 해당 오류 메시지를 출력
    #as e: 해당 구문(Exception)을 축약하여 지정한 매개변수(e) 사용
    print("e")
# 앞서 모든 구문이 끝나고 최종적으로 작동하는 구문
finally:
    print(data14)
```
